## 複習一下 requests 的使用
 
從政府資料開放平台抓資料：https://data.gov.tw/

### 練習

抓取`年度國內主要金融指標`下來做分析

https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp

````python
import requests
````

````python
url = "https://apiservice.mol.gov.tw/OdService/download/A17000000J-030244-dpp"
````

````python
data = requests.get(url).json()
````

````python
# 列出 data 裡面所有的東西
data
````

````python
# 檢視data的資料型態
type(data)
````

````python
# 減試一下 data 裡面有幾筆資料
len(data)
````

````python
# 檢視一下 data 的第一筆資料
data[0]
````

### 列出各年度的M2貨幣供給額


#### 方法1
\

````python
# 利用迴圈列出各年度的M2貨幣供給額
for i in data:
  print(i['年度'], "==>", i['貨幣供給額期底年增率-M2'])
````

#### 方法2

````python
for i in range(len(data)):
  print(i)
````

````python
for i in range(len(data)):
 print(data[i]['年度'], "==>", data[i]['貨幣供給額期底年增率-M2'])
````
